```python
import json
import requests
import pandas as pd
import io
import time
import matplotlib.pyplot as plt
import seaborn as sns
import requests
from bs4 import BeautifulSoup
from typing import List



data = pd.read_json('/content/sample_data/Book.json')
print(len(data))
# print(data)
# print(data.columns)

result = []

for index, row in data.iterrows():
  name = row['Name']
  company = row['Company']
  role = row['Role']
  date = row['Date']
  internshipDuration= row['InternshipDuration']
  intake = row['Intake']

  # print(company," ----> ", role," ----> ", internshipDuration)





  data_ap = {
      'name': name,
      'company': company,
      'role': role,
      'date': date,
      'internshipDuration': internshipDuration,
      'intake': intake,
  }
  result.append(data_ap)
  # time.sleep(1)
print(result)

# def data_sort(package):
#   return package[company][role]


# data.sort(key = data_sort, reverse = True)

# data_str = json.dumps(data_ap, indent= 2)
# print(data_str)




```

```python
from itertools import count
# Create a dictionary to store the counts of each role
role_counts = {}

# Loop through the data and count the roles
for item in result:
    role = item['role']
    if role in role_counts:
        role_counts[role] += 1
    else:
        role_counts[role] = 1

# Sort the roles by count in descending order and print the top 10
sorted_roles = sorted(role_counts.items(), key=lambda x: x[1], reverse=True)
for i, (role, count) in enumerate(sorted_roles[:10]):
    print(f'{i+1}. {role}: {count}')


# Extract the role names and counts for the bar chart
roles = [role for role, count in sorted_roles[:10]]
counts = [count for role, count in sorted_roles[:10]]

# Create a bar chart of the top 10 roles
plt.bar(roles, counts)
plt.title('Top 10 Roles')
plt.xlabel('Role')
plt.ylabel('Count')
plt.show()

# for role,count in role_counts.items():
#   print(f"{role}:{count}")
  
# # Find the role with the most number of occurrences
# most_common_role = max(role_counts, key=role_counts.get)

# # Print the result
# print(f"The most common role is {most_common_role} with {role_counts[most_common_role]} occurrences.")
```
